# Calculate distances between each of the variants and their parent

```python
from sciutil import SciUtil

u = SciUtil()

parent_name = 'H2NOH_G0'
variant_name = 'H2NOH_G7'
def get_hamming(df, parent_name, variant_name):
    parent = df[df['name'] == parent_name]['AA'].values[0]
    variant = df[df['name'] == variant_name]['AA'].values[0]
    len(parent), len(variant)
    # Since the same we can do hamming distance
    hamming_dist = 0
    variant_changes = []
    for i in range(0, len(parent)):
        if parent[i] != variant[i]:
            variant_changes.append(f'{parent[i]}{i+1}{variant[i]}')
            hamming_dist += 1
    u.dp([f'Hamming distance between {parent_name} and {variant_name}', hamming_dist, '_'.join(variant_changes)])
    return variant_changes

get_hamming(lineage_df, 'H2NOH_G0', 'H2NOH_G7')
get_hamming(lineage_df, 'NOPiv_G0', 'NOPiv_G16-2')
get_hamming(lineage_df, 'NOPiv_G0', 'NOPiv_G11-1')
```

```python
import numpy as np
from tqdm import tqdm

# Amino acid code conversion
AA_DICT = {
    "Ala": "A",
    "Cys": "C",
    "Asp": "D",
    "Glu": "E",
    "Phe": "F",
    "Gly": "G",
    "His": "H",
    "Ile": "I",
    "Lys": "K",
    "Leu": "L",
    "Met": "M",
    "Asn": "N",
    "Pro": "P",
    "Gln": "Q",
    "Arg": "R",
    "Ser": "S",
    "Thr": "T",
    "Val": "V",
    "Trp": "W",
    "Tyr": "Y",
    "Ter": "*",
}


def normalise_calculate_stats(processed_plate_df, value_column, normalise="standard", stats_method="mannwhitneyu", parent_label="#PARENT#", normalise_method="median", parent_to_name = None):
    # if nomrliase normalize with standard normalisation
    normalised_df = pd.DataFrame()
    all_stats_df = pd.DataFrame()
    for plate in set(processed_plate_df["barcode_plate"].values):
        # Normalize for each value
        plate_parent = None
        plate_campaign = None
        sub_df = processed_plate_df[processed_plate_df["barcode_plate"] == plate].copy()
        parent_values = sub_df[sub_df["amino_acid_substitutions"] == parent_label][value_column].values
        if len(parent_values) == 0:
            u.err_p(['No parent!!! skipping... '])
            parent_values = sub_df[value_column].values
            continue
        else:
            # By default use the median
            if normalise_method == "median":
                parent_mean = np.nanmedian(parent_values)
            else:
                parent_mean = np.nanmean(parent_values)
            parent_sd = np.nanstd(parent_values)
    
            # For each plate we normalise to the parent of that plate
            sub_df[f"{value_column} plate standard norm"] = (sub_df[value_column].values - parent_mean) / parent_sd
            norm_value_column = f"{value_column} plate standard norm"    
            sd_cutoff = 1.5  # The number of standard deviations we want above the parent values
            # Now for all the other mutations we want to look if they are significant, first we'll look at combinations and then individually
            grouped_by_mutations = sub_df.groupby("amino_acid_substitutions")
        
            rows = []
            for mutation, grp in grouped_by_mutations:
                # Get the values and then do a ranksum test
                if mutation != parent_label:
                    vals = list(grp[norm_value_column].values)
                    U1, p = None, None
                    # Now check if there are 3 otherwise we just do > X S.D over - won't be sig anyway.
                    if len(grp) > 2:
                        # Do stats
                        U1, p = mannwhitneyu(parent_values, vals, method="exact")
                    if normalise_method == "median":
                        mean_vals = np.nanmedian(vals)
                    else:
                        mean_vals = np.nanmean(vals)
                    std_vals = np.nanstd(vals)
                    median_vals = np.nanmedian(vals)
                    sig = mean_vals > ((sd_cutoff * parent_sd) + parent_mean)
                    rows.append([value_column, mutation, len(grp), mean_vals, std_vals, median_vals, mean_vals, sig, U1, p, plate_campaign, plate_parent, plate])
            stats_df = pd.DataFrame(rows, columns=["value_column",  "amino_acid_substitutions", "number of wells with amino_acid substitutions", "mean",
                    "std", "median", "amount greater than parent mean", f"greater than > {sd_cutoff} parent", "man whitney U stat",
                    "p-value", 'campaign',  'parent', 'plate'],
                )
            all_stats_df = pd.concat([all_stats_df, stats_df])
            normalised_df = pd.concat([normalised_df, sub_df])

    return all_stats_df, normalised_df
```

```python
from Bio import AlignIO
from collections import defaultdict, Counter

# Load your alignment (can be fasta, clustal, etc.)
alignment = AlignIO.read("variants_with_larger_than_parent.fasta", "fasta")

groups = {
    "Group1": group1_ids,
    "Group2": group2_ids,
    "Group3": group3_ids,
}

# Create reverse lookup: ID -> group name
id_to_group = {seq_id: group for group, ids in groups.items() for seq_id in ids}

# 1. Report variation per column
print("Column variation report:")
for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    counts = Counter(column)
    variation = len(counts)
    print(f"Position {i+1}: {dict(counts)} (Variation: {variation})")

# 2. Shared and unique mutations across groups
def get_group_column_profiles(group_ids):
    """Returns a list of column profiles for a group."""
    group_seqs = [record.seq for record in alignment if record.id in group_ids]
    profiles = []
    for i in range(alignment.get_alignment_length()):
        residues = tuple(seq[i] for seq in group_seqs)
        profiles.append(residues)
    return profiles

# Get profiles per group
profiles_by_group = {group: get_group_column_profiles(ids) for group, ids in groups.items()}

# Analyze shared and unique mutations
print("\nShared and Unique Mutations:")
for pos in range(alignment.get_alignment_length()):
    residues = {group: profiles_by_group[group][pos] for group in groups}
    
    unique = {}
    for g in groups:
        other_groups = [residues[og] for og in groups if og != g]
        if all(residues[g] != og_res for og_res in other_groups):
            unique[g] = residues[g]
    
    shared = []
    values = list(residues.items())
    for i in range(len(values)):
        for j in range(i+1, len(values)):
            if values[i][1] == values[j][1]:
                shared.append((values[i][0], values[j][0], values[i][1]))

    if unique or shared:
        print(f"Position {pos+1}:")
        if unique:
            for g, v in unique.items():
                print(f"  Unique to {g}: {v}")
        if shared:
            for g1, g2, v in shared:
                print(f"  Shared between {g1} and {g2}: {v}")
```

```python
from Bio import AlignIO
from collections import defaultdict, Counter
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Load alignment
alignment = AlignIO.read("double_variants_with_larger_than_parent.fasta", "fasta")

# Define your group mappings
group1_ids = set(list(positive_df[positive_df['campaign_name'].isin(nopiv_1)]['name'].values) + nopiv_1)
group2_ids = set(list(positive_df[positive_df['campaign_name'].isin(nopiv_2)]['name'].values) + nopiv_2)
group3_ids = set(list(positive_df[positive_df['campaign_name'].isin(h2oh)]['name'].values) + h2oh)

group_map = {}
for i in alignment:
    if i.id in group1_ids:
        group_map[i.id] = "NOPiV_1"
    elif i.id in group2_ids:
        group_map[i.id] = "NOPiV_2"
    elif i.id in group3_ids:
        group_map[i.id] = "H2NOH"

# Updated group_colors with special combinations
group_colors = {
    "NOPiV_1": "#F49CA0",           # blue
    "NOPiV_2": "#C05A6E",           # green
    "H2NOH": "#F39C5A",             # red
    "NOPiV_1+H2NOH": "#F2D95C",     # purple
    "NOPiV_1+NOPiV_2": "#8CC68A",   # orange
    "Shared": "#C3B1E1",               # general shared
    "None": "#AFAFAF",              # white
    "final_H2NOH": '#4EB7A8',
     "final_NOPiV_2": '#3F80D1',
    "final_NOPiV_1": '#A089DB'
}

aa_list = list("ACDEFGHIKLMNPQRSTVWY")
L = alignment.get_alignment_length()

# Compute consensus
consensus = "MTPSDIPGYDYGRVEKSPITDLEFDLLKKTVMLGEEDVMYLKKAADVLKDQVDEILDLAGGWVASNEHLIYYFSNPDTGEPIKEYLERVRARAGAWVLDTTCRDYNREWLDYQYEVGLRHHRSKKGVTDGVRTVPNTPLRYLIAGIYPITATIKPLLAEKGGSPEDIEGMYNAWLKSVVLQVAIWSHPYTKENDWLEHHHHHH"

final_NOPiV_1 = "MTPSDIPGYDYGRVEKSPITDLEFDLLKKTVMLGEEDMMYLKMAADVLKDQVDEILDLVGGWVASNEHLIYYFSNPDTGEPIKEYCERIRARIGAWVLDTTCRDYNREWLDYQYEVGLRHHRSKKGLTDGVRTVPNTPLRYLIAGIYPLTATIKPLLAEKGGSPEDIEGMYNAWLKSVVLQVAIWSHPYTKENDWLEHHHHHH"
final_NOPiV_2 = "MTPSDIPGYDYGRVEKSPITDLEFDLLKKTVMLGEEDIMYLKKAADVLKDQVDEVADLVGGVVASNEHLIYYFSNPDTGEPIKEYCERITARFGAWVLDTTCRDYNREWLDYQYEVGLRHHRSKKGVTDGVRTVPNTPLRYLIAGIYPLTATIKPLLAEKGGSPEDIEGMYNAWLKSMVLQVAIWSHPYTKENDWPEHHHHHH"
final_H2NOH = "MAPSDIPGYDYGRVEKSPITDLEFDLLKTTVMLGEEDVMYLKKAADVLKDQVDEILDLMGRYVASNEHLIYYFSNPDTGEPIKEYFERVRARWGAWVLDTTCRDYNREWLDYQYERGLRHHRSKKGVTDGVRTAPNTPLRYLIAGIYPLTATIKPLLAEKGGSPEDVEGMYNALLKSVVLQVAIWSHPYTKENDWLEHHHHHH"

# Initialize matrix
group_matrix = pd.DataFrame("None", index=aa_list, columns=range(L))

# Assign group labels to amino acids
for pos in range(L):
    consensus_aa = consensus[pos]
    final_NOPiV_1_aa = final_NOPiV_1[pos]
    final_NOPiV_2_aa = final_NOPiV_2[pos]
    final_H2NOH_aa = final_H2NOH[pos]
    
    aa_to_groups = defaultdict(set)
    
    for rec in alignment:
        aa = rec.seq[pos]
        group = group_map.get(rec.id)
        if group:
            aa_to_groups[aa].add(group)
        else:
            aa_to_groups[aa].add('None')
    for aa, groups_here in aa_to_groups.items():
        if aa != consensus_aa:
            if groups_here == {"NOPiV_1", "H2NOH"}:
                label = "NOPiV_1+H2NOH"
            elif groups_here == {"NOPiV_1", "NOPiV_2"}:
                label = "NOPiV_1+NOPiV_2"
            elif len(groups_here) == 1:
                label = list(groups_here)[0]
            else:
                label = "Shared"
            # Check if it's actually in the final variants
            if aa == final_H2NOH_aa:
                label = 'final_H2NOH'
            elif aa == final_NOPiV_2_aa:
                label = 'final_NOPiV_2'
            elif aa == final_NOPiV_1_aa:
                label = 'final_NOPiV_1'
            group_matrix.loc[aa, pos] = label
# Map group labels to colors
color_matrix = group_matrix.applymap(lambda x: group_colors.get(x, "#ffffff"))

# Plotting
plt.figure(figsize=(30, 6))
sns.heatmap(
    pd.DataFrame(np.ones_like(group_matrix.values, dtype=float), index=group_matrix.index),
    annot=False,
    cbar=False,
    cmap=list(group_colors.values()),
    square=False,
    linewidths=0.01,
    linecolor='lightgray',
    mask=(group_matrix == "None").values,
    xticklabels=True,
    yticklabels=True,
)

# Manually color the rectangles
for y in range(group_matrix.shape[0]):
    for x in range(group_matrix.shape[1]):
        group_label = group_matrix.iloc[y, x]
        if group_label != "None":
            plt.gca().add_patch(plt.Rectangle((x, y), 1, 1, color=group_colors[group_label]))

# Legend
legend_elements = [
    Patch(facecolor=color, edgecolor='black', label=label)
    for label, color in group_colors.items() if label != "None"
]
plt.legend(
    handles=legend_elements,
    title="Mutation Group",
    bbox_to_anchor=(1.01, 1),
    loc='upper left',
    borderaxespad=0.
)

plt.title("Final variants")
plt.xlabel("Alignment Position")
plt.ylabel("Amino Acid")
plt.tight_layout()
plt.savefig('final_variants.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from Bio.Align import MultipleSeqAlignment
from Bio import AlignIO
from collections import Counter
from math import log2

def shannon_entropy(column):
    freqs = Counter(column)
    total = sum(freqs.values())
    return -sum((count/total) * log2(count/total) for count in freqs.values())

alignment = AlignIO.read("variants_with_larger_than_parent.fasta", "fasta")
entropies = [shannon_entropy(alignment[:, i]) for i in range(alignment.get_alignment_length())]

plt.figure(figsize=(12, 4))
plt.plot(entropies, color='darkblue')
plt.title("Shannon Entropy Per Alignment Position")
plt.xlabel("Position")
plt.ylabel("Entropy")
plt.tight_layout()
plt.show()
```
